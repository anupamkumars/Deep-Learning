```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# for better accuracy of machine learning models we can also do scaling here like:
# X_train = X_train / 255
# X_test = X_test / 255
# but we have not done any scaling here.
# after scaling when we print X_train[0] then it will print or scale every value between 0 and 1.
```

```python
X_train.reshape(len(X_train),28*28)
```

```python
X_train_flattened = X_train.reshape(len(X_train),28*28)
X_train_flattened.shape

# Flattening is the process of converting a two dimensional array into one dimensional array.
```

```python
X_test_flattened = X_test.reshape(len(X_test),28*28)
X_test_flattened.shape
```

```python
X_train_flattened[0]

# This is a one dimensional array.
```

```python
model = keras.Sequential([
                          keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)


# Sequencial means I have a stack of layers in my neural network.
# Dense means all the neurons in one layer is connected with every other neuron in the second layer.
# Here 10 is output size because we have only 10 neurons in the output layer.
# fit is where training actually happens.
# epoch is the number of iteration for which our neural network is going to run the training.
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[0])

#here my first image is 7
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]

# this will print 10 scores for our 10 output in the array.
```

```python
# Now we need to look the score which is maximum in the above array of 10 scores.
# below function of numpy will find the maximum score and print the index of maximum score.

np.argmax(y_predicted[0])
```

```python
plt.matshow(X_test[1])
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[1]
```

```python
np.argmax(y_predicted[1])
```

#for confusion matrix

```python
y_test[:5]

#truth label
# these are first 5 truth data.
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]

# this will give first 5 predictions.
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions= y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

#adding hidden layer

```python
model = keras.Sequential([
                          keras.layers.Dense(100, input_shape=(784,),activation='relu'),
                          keras.layers.Dense(10, activation='sigmoid')
])

# here I assumed that the number of neurons presented in hidden layer = 100
# here I have added only one hidden layer...but we can add more too
# here 10 is output size.

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```
